```python
import pandas as pd
from faker import Faker
import random
```

```python
fake = Faker()

def generar_agenda(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'fecha': [fake.date_this_year() for _ in range(num_registros)],
        'hora': [fake.time() for _ in range(num_registros)],
        'especialista_id': [random.randint(1, 10) for _ in range(num_registros)],  # Asumiendo 10 especialistas
    }

def generar_cita(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'agenda_id': [random.randint(1, 10) for _ in range(num_registros)],
        'cliente_id': [random.randint(1, 10) for _ in range(num_registros)],
        'estado': [random.choice(['Pendiente', 'Realizada', 'Cancelada']) for _ in range(num_registros)],
    }

def generar_cita_realizada(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'cita_id': [random.randint(1, 10) for _ in range(num_registros)],
        'fecha_realizacion': [fake.date_this_year() for _ in range(num_registros)],
        'observaciones': [fake.text(max_nb_chars=100) for _ in range(num_registros)],
    }

def generar_cliente(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'nombre': [fake.name() for _ in range(num_registros)],
        'telefono': [fake.phone_number() for _ in range(num_registros)],
        'correo': [fake.email() for _ in range(num_registros)],
    }

def generar_especialista(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'nombre': [fake.name() for _ in range(num_registros)],
        'especialidad': [random.choice(['Dentista', 'Cardiologo', 'Pediatra']) for _ in range(num_registros)],
    }

def generar_inventario(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'producto_id': [random.randint(1, 10) for _ in range(num_registros)],
        'cantidad': [random.randint(1, 100) for _ in range(num_registros)],
    }

def generar_procedimiento_servicio(num_registros):
    procedimientos = [
        'Depilación laser',
        'Limpieza facial profunda',
        'Peeling químico',
        'Masaje de reducción y tonificación',
        'Ácido Hialurónico para Rejuvenecimiento Facial',
        'Aumento de Labios con Ácido Hialurónico'
    ]

    costos = {
        'Depilación laser': random.uniform(150000, 350000),
        'Limpieza facial profunda': 140000,
        'Peeling químico': 450000,
        'Masaje de reducción y tonificación': 160000,
        'Ácido Hialurónico para Rejuvenecimiento Facial': 1200000,
        'Aumento de Labios con Ácido Hialurónico': 1400000,
    }

    nombres_procedimientos = [random.choice(procedimientos) for _ in range(num_registros)]
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'nombre': nombres_procedimientos,
        'costo': [costos[nombre] for nombre in nombres_procedimientos],
    }

def generar_producto(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'nombre': [fake.word() for _ in range(num_registros)],
        'precio': [random.uniform(5.0, 200.0) for _ in range(num_registros)],
    }

def generar_proveedor(num_registros):
    return {
        'id': [i for i in range(1, num_registros + 1)],
        'nombre': [fake.company() for _ in range(num_registros)],
        'contacto': [fake.name() for _ in range(num_registros)],
    }

# Número de registros a generar
num_registros = 10

# Generar datos ficticios
df_agenda = pd.DataFrame(generar_agenda(num_registros))
df_cita = pd.DataFrame(generar_cita(num_registros))
df_cita_realizada = pd.DataFrame(generar_cita_realizada(num_registros))
df_cliente = pd.DataFrame(generar_cliente(num_registros))
df_especialista = pd.DataFrame(generar_especialista(num_registros))
df_inventario = pd.DataFrame(generar_inventario(num_registros))
df_procedimiento_servicio = pd.DataFrame(generar_procedimiento_servicio(num_registros))
df_producto = pd.DataFrame(generar_producto(num_registros))
df_proveedor = pd.DataFrame(generar_proveedor(num_registros))

# Mostrar datos generados
print("agenda")
print("")
print(df_agenda.head())
print("cita")
print("")
print(df_cita.head())
print("cita_realizada")
print("")
print(df_cita_realizada.head())
print("cliente")
print("")
print(df_cliente.head())
print("especialista")
print("")
print(df_especialista.head())
print("inventario")
print("")
print(df_inventario.head())
print("procedimiento_servicio")
print("")
print(df_procedimiento_servicio.head())
print("producto")
print("")
print(df_producto.head())
print("proveedor")
print("")
print(df_proveedor.head())
```
